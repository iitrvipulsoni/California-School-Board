```python
import pandas as pd
```

```python
file1 = pd.read_excel("E:/BusyQA/CO-OP 2 - California School Board/Data/Suspension Data.xlsx") 
file2 = pd.read_excel("E:/BusyQA/CO-OP 2 - California School Board/Data/Expulsion Data.xlsx")
file3 = pd.read_excel("E:/BusyQA/CO-OP 2 - California School Board/Data/Chronic Absent Data.xlsx")
file4 = pd.read_excel("E:/BusyQA/CO-OP 2 - California School Board/Data/FRPM Data.xls")
file5 = pd.read_excel("E:/BusyQA/CO-OP 2 - California School Board/Data/Foster Youth Data.xlsx")
```

```python
file12 = pd.merge(file1, file2,  how='inner', left_on=['DistrictName', 'SchoolName','CharterYN', 'ReportingCategory'], right_on = ['DistrictName', 'SchoolName','CharterYN', 'ReportingCategory'])
file12
```

```python
file123 = pd.merge(file12, file3,  how='inner', left_on=['DistrictName', 'SchoolName','CharterYN', 'ReportingCategory'], right_on = ['DistrictName', 'SchoolName','CharterYN', 'ReportingCategory'])
file123
```

```python
file1234 = pd.merge(file123, file4,  how='inner', left_on=['DistrictName', 'SchoolName'], right_on = ['District Name', 'School Name'])
file1234
```

```python
file12345 = pd.merge(file1234, file5,  how='inner', left_on=['DistrictName', 'ReportingCategory'], right_on = ['District Name', 'Student Group Reporting Category'])
file12345
```
